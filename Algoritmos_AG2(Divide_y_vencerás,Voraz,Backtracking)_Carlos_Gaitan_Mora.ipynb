```python
#N Reinas - Vuelta Atrs()
################################################################################

#Verifica que en la solucin parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solucin tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solucin al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(8,solucion=[],etapa=0)
```

## Decorador python para medir el tiempo de ejecucin



```python
import time

# Decorador para medir el tiempo de ejecucin
def medir_tiempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.perf_counter()
        resultado = func(*args, **kwargs)
        fin = time.perf_counter()
        print(f"Tiempo de ejecucin de '{func.__name__}': {fin - inicio:.6f} segundos")
        return resultado
    return wrapper
```

## Funcion de distancia entre puntos
Se usara la formula para la distancia euclidiana (ref: https://en.wikipedia.org/wiki/Euclidean_distance)

```python
import numpy as np

def distancia_euclidiana(punto_uno, punto_dos):
  """
  Calcula la distancia entre los dos puntos dados.
  IMPORTANTE: ambos puntos deben estar en la misma dimension.

  Parmetros:
  - punto_uno: Punto de partida puede estar en 1, 2 o 3 dimenciones,
    no puede ser nulo y solo puede ser de tipo flotante, entero o tupla.
  - punto_dos: Punto de destino puede estar en 1, 2 o 3 dimenciones,
    no puede ser nulo y solo puede ser de tipo flotante, entero o tupla.

  Retorna:
  - La distancia euclidiana entre los dos puntos en un flotante
  """
  p1 = None
  p2 = None
  # primero validamos en que dimensiones esta los puntos
  if isinstance(punto_uno,  float) or isinstance(punto_uno,  int):
    # si es una dimension simplemente restamos y devolvemos el absoluto
    return float(abs(punto_uno - punto_dos))
  else:
    # si esta en 2 o mas dimensiones aplicamos la formula
    p1 = np.array(punto_uno)
    p2 = np.array(punto_dos)
    # Complejidad de esto es lineal, dado que depende de la dimension de los
    # puntos pero nuestro peor escenario en este ejercicio es 3 osea
    # O(n) -> O(3), lo cual es despreciable.
    resta_y_potencia = np.sum((p1-p2)**2, axis=0)
    return np.sqrt(resta_y_potencia)

# pequea prueba del metodo
print(distancia_euclidiana(1, 2))
print(distancia_euclidiana((1,1), (1,2)))
print(distancia_euclidiana((1,1,1), (1,1,2)))
```

## Primer intento por fuerza bruta

```python
@medir_tiempo
def fuerza_bruta(puntos: list):
  """
  Calcula la distancia entre los dos puntos dados.
  IMPORTANTE: ambos puntos deben estar en la misma dimension.

  Parmetros:
  - puntos: lista con los puntos a revisar, puede ser en una dimension, dos
    dimensiones o hasta tres dimensiones.

  Retorna:
  - Una tupla con:
    La distancia euclidiana entre los dos puntos en un flotante
    El punto de partida con el que se encontro la distancia mas corta
    El punto de destino con el que se encontro la distancia mas corta
    La posicion en la lista del punto de partida
    La posicion en la lista del punto de destino
  """
  min_distancia = 99999
  p1 = None
  p2 = None
  pos_p1 = None
  pos_p2 = None
  total_puntos = len(puntos)
  # Calculamos el punto contra todos los siguientes
  # Complejidad cuadratica n*m [n =  len(puntos-1), m = len(puntos-actual)]
  for i in range(total_puntos - 1):
    for j in range(i+1, total_puntos):
      distancia_actual = distancia_euclidiana(puntos[i], puntos[j])
      if distancia_actual < min_distancia:
        min_distancia = distancia_actual
        p1 = puntos[i]
        p2 = puntos[j]
        pos_p1 = i
        pos_p2 = j
  return (min_distancia, p1, p2, pos_p1, pos_p2)

min, p1, p2, pos_p1, pos_p2 = fuerza_bruta(lista_1d)
print(f"Resultado 1D, distancia mas corta: {min}")
print(f"    punto 1: {p1}, punto 2: {p2}")
print(f"    posicin del punto 1: {pos_p1}, posicin del punto 2: {pos_p2}")

min, p1, p2, pos_p1, pos_p2 = fuerza_bruta(lista_2d)
print(f"Resultado 2D, distancia mas corta: {min}")
print(f"    punto 1: {p1}, punto 2: {p2}")
print(f"    posicin del punto 1: {pos_p1}, posicin del punto 2: {pos_p2}")

min, p1, p2, pos_p1, pos_p2 = fuerza_bruta(lista_3d)
print(f"Resultado 3D, distancia mas corta: {min}")
print(f"    punto 1: {p1}, punto 2: {p2}")
print(f"    posicin del punto 1: {pos_p1}, posicin del punto 2: {pos_p2}")
```

### Complejidad del algoritmo

n = total de puntos

m = dimension de los puntos

$$ O(m) * O(n^{2})$$

Eliminando las constantes y despreciando el clculo de la distancia euclidiana, ya que para cada ciclo es $O(3)$, en el peor escenario nos quedamos con:

$$O(n^{2})$$

Dado que para obtener la distancia menor es necesario comparar todos los puntos, quizs podramos aplicar Quicksort, que tiene una complejidad de $O(n*log(n))$ y validar los puntos que creeria es lineal dado que solo es revisar de a pares, pero creo que en el peor escenario, la complejidad seguira siendo cuadrtica $O(n^{2})$

## Segundo intento por divide y venceras

```python
def divide_y_venceras(puntos: list, ordenados = False):
  """
  Calcula la distancia entre los dos puntos dados.
  IMPORTANTE: ambos puntos deben estar en la misma dimension.

  Parmetros:
  - punto_uno: Punto de partida puede estar en 1, 2 o 3 dimenciones,
    no puede ser nulo y solo puede ser de tipo flotante, entero o tupla.
  - punto_dos: Punto de destino puede estar en 1, 2 o 3 dimenciones,
    no puede ser nulo y solo puede ser de tipo flotante, entero o tupla.

  Retorna:
  - La distancia euclidiana entre los dos puntos en un flotante
  """
  min_distancia = 99999
  p1 = None
  p2 = None
  pos_p1 = None
  pos_p2 = None
  total_puntos = len(puntos)
  puntos_odenados = puntos
  rango_vecinos = 10

  # Ordenamos la lista usando la distancia al punto 0
  # Complejidad n*log(n) dado que usamos quicksort, sin embargo en el peor
  # escenario puede ser cuadratica.
  if not ordenados:
    punto_ref = puntos[0]
    punto_zero = [0]*len(puntos[0]) if isinstance(punto_ref, tuple) else 0
    puntos_odenados = sorted(puntos, key= lambda x: distancia_euclidiana(punto_zero, x))

  # partimos la lista a la mitad hasta que sean 2 puntos a evaluar y tomamos
  # el menor del calculo.
  # Complejidad n dado que al final recorremos los puntos por cierto rango
  # minimo y no evaluamos todos contra todos.
  if total_puntos > rango_vecinos:
    pivote = total_puntos // 2
    mitad_izquierda = puntos_odenados[:pivote + 1]
    mitad_derecha = puntos_odenados[pivote:]

    distancia_i, p_i1, p_i2 = divide_y_venceras(mitad_izquierda, True)
    distancia_d, p_d1, p_d2 = divide_y_venceras(mitad_derecha, True)

    if distancia_d <= distancia_i:
      return (distancia_d, p_d1, p_d2)
    else:
      return (distancia_i, p_i1, p_i2)
  else:
    # si la lista esta dentro del rango de vecinos a revisar solo resta
    # sacar los puntos con la distancia minima.
    for i in range(total_puntos - 1):
      for j in range(i+1, total_puntos):
        distancia_actual = distancia_euclidiana(puntos[i], puntos[j])
        if distancia_actual < min_distancia:
          min_distancia = distancia_actual
          p1 = puntos[i]
          p2 = puntos[j]
    return min_distancia, p1, p2

@medir_tiempo
def divide_y_venceras_tiempo(puntos: list):
  return divide_y_venceras(puntos)

min, p1, p2 = divide_y_venceras_tiempo(lista_1d)
print(f"Resultado 1D, distancia mas corta: {min}")
print(f"    punto 1: {p1}, punto 2: {p2}")
min, p1, p2 = divide_y_venceras_tiempo(lista_2d)
print(f"Resultado 2D, distancia mas corta: {min}")
print(f"    punto 1: {p1}, punto 2: {p2}")
min, p1, p2 = divide_y_venceras_tiempo(lista_3d)
print(f"Resultado 3D, distancia mas corta: {min}")
print(f"    punto 1: {p1}, punto 2: {p2}")
```

### Complejidad del algoritmo

n = total de puntos

m = dimension de los puntos

$$O(m) * O(n*log(n)) + O(n)$$

Eliminando constantes y despreciando el clculo de la distancia euclidiana nuevamente, ya que en cada ciclo es $O(3)$ en el peor escenario, nos quedamos con:

$$O(n*log(n))$$

Sin embargo, dado que se utiliza Quicksort, es importante recordar que, en el peor escenario, la complejidad puede seguir siendo:

$$O(n^{2})$$
